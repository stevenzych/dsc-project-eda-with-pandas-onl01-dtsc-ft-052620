```python
# Let's get started importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading the data
df = pd.read_csv("ames_train.csv")
df.head()
```

```python
# Investigate the Data
df.info()
```

```python
# Investigating Distributions using scatter_matrix
pd.plotting.scatter_matrix(df[["SalePrice", "TotalBsmtSF"]])
```

```python
# Create a plot that shows the SalesPrice Distribution
fig, sales = plt.subplots(figsize = (10,8))
sales.hist(df["SalePrice"], bins = "auto")
plt.show()
```

```python
# Create a plot that shows the LotArea Distribution
fig, lots = plt.subplots(figsize = (10,8))
lots.hist(df["LotArea"], bins = "auto")
plt.show()
```

```python
# Create a plot that shows the Distribution of the overall house condition
fig, cond = plt.subplots(figsize = (10,8))
cond.hist(df["OverallCond"], bins = "auto")
cond.set_xlabel("Condition of House", fontsize = 20)
cond.set_ylabel("Number of Houses", fontsize = 20)
plt.show()
```

```python
# Create a Box Plot for SalePrice
fig, spb = plt.subplots(figsize = (10,8))
spb.boxplot(df["SalePrice"])
spb.set_ylabel("Sale Price", fontsize = 20)
plt.show()
```

```python
# Perform an Exploration of home values by age
pd.plotting.scatter_matrix(df[["SalePrice", "YearBuilt"]])
```
